# Covarianza

## Definición.

La covarianza cuantifica la fuerza y la dirección de una relación entre dos variables. Es el dato básico para determinar si existe una dependencia entre ambas variables y además es el dato necesario para estimar otros parámetros básicos, como el coeficiente de correlación lineal o la recta de regresión. 

<div class="alert alert-block alert-info">

**Definición.**

Dados los siguientes conjuntos de datos $X = \{x_1, \dots, x_N\}$ y $Y = \{y_1, \dots, y_N\}$ la covarianza entre las variables $X$ y $Y$ se calcula como sigue:

$$
S_{xy} = \frac{1}{N-1} \sum_{i=1}^{N} (x_i - \mu_{x})(y_i - \mu_{y})
$$

donde $\mu_{x}$ y $\mu_{y}$ representan la media de $X$ y $Y$ respectivamente.

* Si $S_{xy}>{0}$ hay dependencia directa (positiva), es decir, a grandes valores de $x$ corresponden grandes valores de $y$.

* Si $S_{xy}={0}$ se interpreta como la no existencia de una relación lineal entre las dos variables.

* Si $S_{xy}<{0}$ hay dependencia inversa (negativa), es decir, a grandes valores de $x$ corresponden pequeños valores de $y$.

## Ejercicio 2.

Calcula la covarianza entre el `grupo_A` y el `grupo_B` usando la siguiente fórmula:

$$
S_{xy} = \frac{1}{N-1} \sum_{i=1}^{N} (x_i - \mu_{x})(y_i - \mu_{y})
$$

Pasos a seguir:
* Primero calcula la media del `grupo_A` y del `grupo_B`, usando la función `media()`, y almacénalas en las variables `μA` y `μB` respectivamente.
* Implementa el ciclo de la suma como sigue (completa el código):

```python
suma = 0.0  
for xi, yi in zip(grupo_A, grupo_B):
    suma += ... 
```
* Finalmente divide la suma entre `N-1` y almacena el resultado en la variable `Sx`.
  
</div>

````python
# Calcular la covarianza entre el grupo_A y el grupo_B
### BEGIN SOLUTION
μx = media(grupo_A)
μy = media(grupo_B)

suma = 0.0  
for xi, yi in zip(grupo_A, grupo_B):
    suma += (xi - μx) * (yi - μy)

Sxy = suma / (N-1)
### END SOLUTION

print(f"Covarianza Sxy = {Sxy:8.4f}")
````

````python
# Función para calcular la covarianza entre dos variables X y Y
### BEGIN SOLUTION
def covarianza(X, Y):
    μx = media(X)
    μy = media(Y)
    
    suma = 0.0  
    for xi, yi in zip(X, Y):
        suma += (xi - μx) * (yi - μy)
    
    return suma / (len(X)-1)
### END SOLUTION
````

La covarianza es una medida importante ya que permite identificar cómo dos variables se comportan en conjunto: si tienden a aumentar o disminuir simultáneamente, o si se mueven en direcciones opuestas. Esta información es clave para entender relaciones subyacentes entre conjuntos de datos, y sirve como base para construir modelos que expliquen cómo una variable responde ante los cambios de otra. Al cuantificar esta relación, se pueden anticipar comportamientos, validar hipótesis y tomar decisiones más informadas en contextos analíticos, predictivos o comparativos.
