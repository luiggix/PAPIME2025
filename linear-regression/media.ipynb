# Media

## Definición

La media es una medida de tendencia central que representa el valor promedio de un conjunto de datos. Para obtener este valor se suman todos los valores del conjunto y se divide entre el número total de valores del conjunto.

Por ejemplo, si tenemos el siguiente conjunto de valores $X = \{3, 5, 1, 2\}$, entonces la media se calcula como sigue: 

$$\mu = (3 + 5 + 1 + 2) / 4 = 2.75$$ 

El símbolo $\mu$ se usa para representar la media.

<div class="alert alert-block alert-info">

**Definición.**

Dado el conjunto de datos $X = \{x_1, \dots, x_N\}$ su media se define como sigue:

$$
\mu = \dfrac{1}{N}\sum_{i=1}^{N} x_i
$$

## Ejercicio 3.
El siguiente código representa una función de Python que calcula la media de un conjunto de valores. La función recibe el arreglo `X` con los valores, calcula la media y regresa su valor. 

```python
def media(X):
    suma = 0
    for xi in X:
        suma += xi
    
    return suma / len(X)
```

La media es una medida importante ya que resume en un solo valor el comportamiento central de un conjunto de datos. Al calcular el promedio, se obtiene una referencia que permite comparar observaciones, detectar desviaciones y entender la tendencia general del fenómeno analizado. Esta medida es esencial para interpretar resultados, establecer puntos de referencia y tomar decisiones basadas en datos. Además, sirve como base para otras métricas estadísticas más complejas, y es clave en modelos que buscan explicar o predecir comportamientos a partir de patrones.
