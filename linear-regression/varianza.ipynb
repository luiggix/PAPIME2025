# Varianza

## Definición

La varianza es una medida estadística que indica qué tan alejados están los valores individuales respecto a la media del conjunto, es decir cuánto se dispersan. Un valor alto de varianza significa que los datos están muy dispersos, mientras que un valor bajo indica que están más agrupados alrededor de la media. La varianza se calcula como la suma de los residuos elevados al cuadrado y divididos entre el total de observaciones. A esta varianza se le conoce como **varianza poblacional**.

<div class="alert alert-block alert-info">

**Definición.**

Dado el conjunto de datos $X = \{x_1, \dots, x_N\}$ su **varianza poblacional** se calcula como  

$$\sigma^2 = \frac{1}{N} \sum_{i=1}^{N} (x_i - μ)^2$$

</div>


<div class="alert alert-block alert-info">

**Definición.**

Dado el conjunto de datos $X = \{x_1, \dots, x_N\}$ su **varianza muestral** se calcula como  

$$s^2 = \frac{1}{N-1} \sum_{i=1}^{N} (x_i - μ)^2$$

Se divide entre (n-1) para estimar la varianza de la población de manera más precisa

</div>

Por ejemplo, si tenemos el siguiente conjunto de valores $X = \{3, 5, 1, 2\}$, entonces la varianza se calcula como sigue: 

* La media es: $\mu = (3 + 5 + 1 + 2) / 4 = 11 / 2.75$. 
* La varianza es:

$$
\begin{eqnarray}
\sigma^2 & = & \Big( (3-2.75)^2 + (5-2.75)^2 \\
& & + (1-2.75)^2 + (2-2.75)^2\Big) / 4 \\
\sigma^2 & = & 2.1875
\end{eqnarray}
$$

## Ejercicio 3.

* Definir la siguiente función para calcular la varianza de un conjunto de datos. Esta función recibe el conjunto de datos `X` y el valor de la media del conjunto `μ`.

```python
def varianza(X, μ):
    suma = 0
    for xi in X:
        suma += (xi - μ)**2
    
    return suma / len(X)
```

* Utilizar la función `varianza()` para calcular la varianza de cada grupo de tal manera que el resultado sea vea como sigue:

```
Varianza grupo A = 21.45956212187354
Varianza grupo B = 76.62578602767512
Varianza grupo C = 50.32562930760074
Varianza grupo D = 162.10093386561326
```
<br>

**Hint**. Primero almacena los valores de la varianza en las variables `σ2A`, `σ2B`, `σ2C` y `σ2D` para cada grupo. Posteriormente usa la función `print()` para imprimir los valores calculados y almacenados en dichas variables.

</div>

````python
# Función para calcular la varianza
### BEGIN SOLUTION
def varianza(X, μ):
    suma = 0
    for xi in X:
        suma += (xi - μ)**2
    
    return suma / len(X)
### END SOLUTION
````

````python
# Cálculo e impresión de la varianza de cada grupo
### BEGIN SOLUTION
σ2A = varianza(grupo_A, μA)
σ2B = varianza(grupo_B, μB)
σ2C = varianza(grupo_C, μC)
σ2D = varianza(grupo_D, μD)

print(f"Varianza grupo A = {σ2A}")
print(f"Varianza grupo B = {σ2B}")
print(f"Varianza grupo C = {σ2C}")
print(f"Varianza grupo D = {σ2D}")
### END SOLUTION
````
